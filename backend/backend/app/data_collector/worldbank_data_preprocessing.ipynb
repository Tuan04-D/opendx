```python
import json
import pandas as pd
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# ===== Load d liu =====
with open("D:\Study\OpenDX\data\worldbank\IT.NET.USER.ZS\IT.NET.USER.ZS.json", "r", encoding="utf-8") as f:
    raw = json.load(f)

data = raw[1]   # phn d liu nm  index 1

# ===== Chuyn vo DataFrame =====
df = pd.DataFrame(data)[["date", "value"]].dropna()

# i kiu d liu cho Prophet
df["ds"] = pd.to_datetime(df["date"])
df["y"] = df["value"].astype(float)
df = df[["ds", "y"]].sort_values("ds")

print(df.tail())  # kim tra d liu cui

```

```python

# ===== Chia train-test theo thi gian =====
train = df[df["ds"] < "2022"]   # train n nm 2021
test = df[df["ds"] >= "2022"]   # test t 2022 tr i

print("Train size:", len(train))
print("Test size:", len(test))

# ===== Train model =====
model = Prophet()
model.fit(train)

# ===== D bo trn tp test =====
future_test = test[["ds"]]
forecast_test = model.predict(future_test)

# ===== nh gi =====
y_true = test["y"].values
y_pred = forecast_test["yhat"].values

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f"\nnh gi m hnh:")
print(f"MAE  = {mae:.3f}")
print(f"RMSE = {rmse:.3f}")
```

```python
# ===== D bo 5 nm tip theo =====
future = model.make_future_dataframe(periods=5, freq="Y")
forecast_future = model.predict(future)

pred_5y = forecast_future[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail(5)
print("\n D bo 5 nm tip theo:")
print(pred_5y)

# ===== Nu mun lu kt qu =====
pred_5y.to_csv("forecast_next_5_years.csv", index=False)
```
